```python
import pandas as pd
import numpy as np

from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
samp_cst_feat = pd.read_csv('data/samp_cst_feat.csv', encoding='ISO-8859-1')
samp_train = pd.read_csv('data/samp_train.csv', encoding='ISO-8859-1')
```

```python
samp_cst_feat.head()
```

```python
df = samp_cst_feat.copy()
df['label'] = samp_train['MRC_ID_DI']
df
```

```python
X = df.drop(['label'], axis = 1)
Y = df['label']
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)
PCA_df = pd.DataFrame(data = X_pca, columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
PCA_df.head()
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(PCA_df, Y, test_size = 0.30, random_state = 101)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0)
X_Over,Y_Over = smote.fit_sample(X_Train, Y_Train)
```

```python
print('Data shape before oversampling: ', X_Train.shape, Y_Train.shape)
print('Data shape after oversampling:: ', X_Over.shape, Y_Over.shape)
print('Label distribution: \n', pd.Series(Y_Over).value_counts())
```
